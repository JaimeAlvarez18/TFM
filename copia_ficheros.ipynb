```python
import os
import cv2
from pyspark.sql import SparkSession

# Set source and destination root directories
source_root = "/media/jaime/DATASETS/Datasets/TFM/GenImage"
dest_root = "/home/jaime/Desktop/GenImage_resized"
os.makedirs(dest_root, exist_ok=True)

# Recursively collect all image file paths and their relative paths
image_files = []
for class_name in os.listdir(source_root):
    class_path = os.path.join(source_root, class_name)
    if os.path.isdir(class_path):
        for split in ["train", "val"]:
            split_path = os.path.join(class_path, split)
            if os.path.isdir(split_path):
                for fname in os.listdir(split_path):
                    if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                        abs_path = os.path.join(split_path, fname)
                        rel_path = os.path.join(class_name, split, fname)
                        image_files.append((abs_path, rel_path))

# Initialize Spark
spark = SparkSession.builder.appName("ImageResizeCopy").getOrCreate()
sc = spark.sparkContext

def resize_and_copy(paths):
    src_path, rel_path = paths
    dest_path = os.path.join(dest_root, rel_path)
    os.makedirs(os.path.dirname(dest_path), exist_ok=True)
    try:
        img = cv2.imread(src_path)
        if img is None:
            return f"Failed to read {src_path}"
        resized = cv2.resize(img, (800, 600))
        cv2.imwrite(dest_path, resized)
        return f"Copied and resized {rel_path}"
    except Exception as e:
        return f"Error processing {src_path}: {e}"

# Parallelize and process images
results = sc.parallelize(image_files).map(resize_and_copy).collect()
for res in results:
    print(res)
```
